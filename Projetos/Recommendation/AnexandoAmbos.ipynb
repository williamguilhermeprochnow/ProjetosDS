```python
df_movie_ratings = pd.read_csv('Arquivos/MovieRatings.csv', index_col='imdb_title_id', parse_dates=['date_published'])
df_movie_ratings.info()
```

```python
df_movie_ratings.head()
```

```python
df_movie_ratings['genre'] = df_movie_ratings['genre'].str.rstrip().apply(lambda x: x.replace(" ", ""))
genre = list(df_movie_ratings['genre'][~df_movie_ratings['genre'].str.contains(',')].unique())
```

```python
genre
```

```python
# Verificar se o genre existe na lista de genero (conatins).
nome = '|'.join(genre)
df_movie_ratings['different_genre'] = df_movie_ratings.genre.str.contains(nome)
df_movie_ratings.query("different_genre == False")
```

```python
df_movie_ratings.drop(columns=['different_genre'], inplace=True)
```

```python
new_cols = genre
df_movie_ratings = df_movie_ratings.assign(**dict.fromkeys(new_cols, 0))
```

```python
df_movie_ratings
```

```python
df_movie_ratings.info()
```

```python
for i in range(0, len(df_movie_ratings)):
    lista = df_movie_ratings['genre'][i].split(',')
    df_movie_ratings.loc[df_movie_ratings.index[i], lista] = int(1)
```

```python
df_movie_ratings
```

```python
df_movie_ratings[df_movie_ratings['Documentary'] == 1]
```

```python
df_movie_ratings[['Film-Noir', 'Documentary', 'Reality-TV', 'News']] = df_movie_ratings[['Film-Noir', 'Documentary', 'Reality-TV', 'News']].fillna(0)
df_movie_ratings[['Film-Noir', 'Documentary', 'Reality-TV', 'News']] = df_movie_ratings[['Film-Noir', 'Documentary', 'Reality-TV', 'News']]\
                                                                       .astype('int64')
```
