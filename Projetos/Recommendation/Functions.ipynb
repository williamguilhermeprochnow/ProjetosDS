```python
def genreEquals(movie):
    genre_list_main = []
    genre_list = []
    movies_genres_equals = []
    
    index_movie = df_movie_ratings.loc[:, 'Romance':][df_movie_ratings['original_title'] == movie].index[0]
    
    
    select = pd.DataFrame(df_movie_ratings.loc[:, 'Romance':][df_movie_ratings['original_title'] == movie]
                                          .values[0]).T
    select.columns = [df_movie_ratings.loc[:, 'Romance':].columns]
    
    for i in range(0, len(select.columns[select.values[0] > 0])):
        genre_list_main.append(select.columns[select.values[0] > 0][i][0])
    genre_list_main.sort()
    
    
    for i in range(0, 1000):
        
        others_movies = pd.DataFrame(df_movie_ratings.loc[:, 'Romance':][df_movie_ratings.index != index_movie]
                                                     .values[i]).T
        others_movies.columns = [df_movie_ratings.loc[:, 'Romance':].columns]
        
        for i_columns in range(0, len(others_movies.columns[others_movies.values[0] > 0])):
            genre_list.append(others_movies.columns[others_movies.values[0] > 0][i_columns][0])
        genre_list.sort()
        
        
        if genre_list_main == genre_list:
            if int(df_movie_ratings[df_movie_ratings['original_title'] == movie].index[0]) < i:
                movies_genres_equals.append(df_movie_ratings['original_title'][i+1])
            else:
                movies_genres_equals.append(df_movie_ratings['original_title'][i])
        genre_list.clear()
                
    movies_genres_equals.sort()
            
    return movies_genres_equals
```

## Início para a tabela principal

### Semelhança entre o gênero dos outros filmes

```python
StartIn = 0
NumberMovieSearch = 100


def genreEquality(moviegenre):
    genre_list_main = []
    genre_list = []
    movies_genres_equals = []
    
    index_movie = df_movie_ratings.loc[:, 'Romance':][df_movie_ratings['original_title'] == moviegenre].index[0]
    
    
    select = pd.DataFrame(df_movie_ratings.loc[:, 'Romance':][df_movie_ratings['original_title'] == moviegenre]
                                          .values[0]).T
    select.columns = [df_movie_ratings.loc[:, 'Romance':].columns]
    
    for i in range(0, len(select.columns[select.values[0] > 0])):
        genre_list_main.append(select.columns[select.values[0] > 0][i][0])
    
    
    for i in range(StartIn, NumberMovieSearch):
        
        others_movies = pd.DataFrame(df_movie_ratings.loc[:, 'Romance':][df_movie_ratings.index != index_movie]
                                                     .values[i]).T
        others_movies.columns = [df_movie_ratings.loc[:, 'Romance':].columns]
        other_movie = df_movie_ratings['original_title'][df_movie_ratings.index != index_movie].values[i]
        
        for i_columns in range(0, len(others_movies.columns[others_movies.values[0] > 0])):
            genre_list.append(others_movies.columns[others_movies.values[0] > 0][i_columns][0])
        
        equals_genres = list(filter(lambda x: x in genre_list_main, genre_list))
        
        if len(genre_list) > 1:
            percentage = len(equals_genres) / len(genre_list)
        else:
            percentage = len(equals_genres) / len(genre_list_main)
            
            
        movies_genres_equals.append([other_movie, percentage])
            
        genre_list.clear()
        equals_genres.clear()
        

    df_equals_genres = pd.DataFrame(movies_genres_equals, columns=['Movie', 'Equity Genre'])
    df_equals_genres['Equity Genre'] = df_equals_genres['Equity Genre'].round(2)
    
    
    return df_equals_genres
```

### Semelhança dentre todos os aspectos

```python
def generalEquality(movie, ordenation):
    columns = df_movie_ratings.columns[[5,6,7,8,9,0]]
    name_columns = ['Country', 'Director', 'Writer', 'Company', 'Actors', 'Title']
    movies_equals = []
    
    df_equals_movies = genreEquality(movie)
    
    for df in range(0, len(columns)):
        movie_list_main = []
        movie_list = []

        index_movie = df_movie_ratings.loc[:, f'{columns[df]}'][df_movie_ratings['original_title'] == movie].index[0]

        select = df_movie_ratings.loc[:, f'{columns[df]}'][df_movie_ratings['original_title'] == movie].values[0]
        movie_list_main.append(select.split(", "))


        for i in range(StartIn, NumberMovieSearch):

            others_movies = df_movie_ratings.loc[:, f'{columns[df]}'][df_movie_ratings.index != index_movie].values[i]
            movie_list.append(others_movies.split(", "))
            other_movie = df_movie_ratings['original_title'][df_movie_ratings.index != index_movie].values[i]

            equals_movies = list(filter(lambda x: x in movie_list_main[0], movie_list[0]))

            if len(movie_list[0]) > 1:
                percentage = len(equals_movies) / len(movie_list[0])
            else:
                percentage = len(equals_movies) / len(movie_list_main[0])

            movies_equals.append([other_movie, percentage])

            movie_list.clear()
            equals_movies.clear()
        
        df_equals_movies[['Movie', f'Equity {name_columns[df]}']] = pd.DataFrame(movies_equals)
        df_equals_movies[f'Equity {name_columns[df]}'] = df_equals_movies[f'Equity {name_columns[df]}'].round(2)
        movies_equals.clear()
        
    df_general_equality = df_equals_movies
    df_general_equality['General (%)'] = df_general_equality.loc[:, 'Equity Genre':].sum(axis=1)
    df_general_equality['General (%)'] = (df_general_equality['General (%)'] / 7).round(2) * 100
    
    try:
        return df_general_equality.sort_values(by=ordenation, ascending=False)
    except:
        return df_general_equality.sort_values(by='General (%)', ascending=False)
```
